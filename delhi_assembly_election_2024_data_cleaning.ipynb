```python
# getting ac name and ac number 

delhi_ae_24[['ac_name','ac_no']]  = delhi_ae_24['Dropdown_Option'].str.split('-',expand = True)
```

```python
#formatting column names
delhi_ae_24.columns = delhi_ae_24.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
delhi_ae_24
```

```python
delhi_ae_24.rename(columns = {'candidate':'candidate_name','total_votes':'total_votes_secured',
                            '%_of_votes':'vote_share_percentage'}, inplace=True)
```

```python
total = delhi_ae_24[delhi_ae_24['candidate_name']=='Total'][['total_votes_secured','ac_no','ac_name']]

total
```

```python
drop_index = delhi_ae_24[delhi_ae_24['candidate_name']=='Total'].index # filtering total
```

```python
delhi_ae_24.drop(index=drop_index,inplace= True) #droping total
```

```python
delhi_ae_24.head(500)
```

```python
# generating ac wise total vote column
delhi_ae_24['total_polled_votes'] = delhi_ae_24.groupby('ac_no')['total_votes_secured'].transform('sum')
```

```python
delhi_ae_24.sort_values(by='total_polled_votes')['total_polled_votes'].drop_duplicates().reset_index(drop='index').values
```

```python
#extracted total votes from ECI and calculated total votes match



total.sort_values(by='total_votes_secured')['total_votes_secured'].reset_index(drop='index').values == delhi_ae_24.sort_values(by='total_polled_votes')['total_polled_votes'].drop_duplicates().reset_index(drop='index').values
```

```python
delhi_ae_24.sort_values(by=['ac_no','total_votes_secured'],ascending = [True,False],inplace = True)
```

```python
delhi_ae_24['rank'] = delhi_ae_24.groupby('ac_no')['total_votes_secured'].rank(method='min',ascending = False)
```

```python
delhi_ae_24.head(3)
```

```python
# matches with the final results, BJP - 48 , AAAP - 22 Total ac seat - 70

delhi_ae_24[delhi_ae_24['rank'].isin([1,2,3])].groupby(['party','rank'])['rank'].count().unstack()
```

```python
delhi_ae_24.drop(columns = ['s.n.','dropdown_option'], inplace = True) # dropping unnecessary columns
```

```python
party_mapper = {'Bharatiya Janata Party':'BJP','Aam Aadmi Party':'AAAP','Indian National Congress':'INC'}
```
